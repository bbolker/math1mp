```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as npr
x = np.linspace(0,1,101)  ## alternative to np.arange()
y = np.sqrt(1-x**2)
```

```python
%matplotlib inline
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
xr = npr.rand(10000)
yr = npr.rand(10000)
incirc = xr**2+yr**2<1
## print(incirc)
```

```python
%matplotlib inline
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(xr[incirc],yr[incirc],"b*")
ax.plot(xr[np.logical_not(incirc)],
         yr[np.logical_not(incirc)],"ro")
```

```python
c = ((2,2.5),(4,3),(1,2))
r = (1.5,3,2)
npr.seed(101)
x = npr.uniform(c[0][0]-r[0],c[0][0]+r[0],size=1000000)
y = npr.uniform(c[0][1]-r[0],c[0][1]+r[0],size=1000000)
tests = np.tile(True,10**6)
def incirc(x,y,ctr,radius):
    dsq = (x-ctr[0])**2+(y-ctr[1])**2
    return(dsq<radius**2)
```

```python
list(zip(c,r))
```

```python
for c0,r0 in zip(c,r):
    tests = tests & incirc(x,y,c0,r0)
    print(np.mean(tests))
```
